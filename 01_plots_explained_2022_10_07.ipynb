```python
#cell 1
#binary distribution (heads vs tails)

import matplotlib.pyplot as plt
import numpy as np

d_min=1
d_max=3
n=50 # sample size

data=np.random.randint(d_min, d_max, n);
fig, axs = plt.subplots(nrows=3, figsize=(12, 10));
axs[0].plot(data, 'o', mfc = 'none',  markersize = 12);
axs[0].set_ylabel("amplitude", fontsize=14);
bins = np.arange(0.0, 3, 0.1)
axs[1].hist(data, bins);
axs[1].set_ylabel("observations", fontsize=14);
axs[2].hist(data, bins, density=False, weights=np.ones(len(data)) / len(data));
axs[2].set_ylabel("proportion", fontsize=14);
```

```python
# cell 2
# uniform distribution 1:2 (e.g cards?)

import matplotlib.pyplot as plt
import numpy as np

d_min = 1
n=500000 # sample size

data=np.random.rand(n)+d_min;
fig, axs = plt.subplots(nrows=3, figsize=(12, 10));
axs[0].plot(data, 'o', mfc = 'none',  markersize = 12);
axs[0].set_ylabel("amplitude", fontsize=14);
bins = np.arange(0.0, 3, 0.1)
axs[1].hist(data, bins);
axs[1].set_ylabel("observations", fontsize=14);
axs[2].hist(data, bins, density=False, weights=np.ones(len(data)) / len(data));
axs[2].set_ylabel("proportion", fontsize=14);
```

```python
# cell 3
# normal distribution, e.g. multiple aimimg a target with a gun

import matplotlib.pyplot as plt
import numpy as np

mu = 1.5 # planned mean
sigma = 0.3 # planned standard deviation
n=50 # sample size

data=np.random.randn(n)*sigma+mu;
fig, axs = plt.subplots(nrows=3, figsize=(12, 10));
axs[0].plot(data, 'o', mfc = 'none',  markersize = 12);
axs[0].set_ylabel("amplitude", fontsize=14);
bins = np.arange(0.0, 3, 0.1)
axs[1].hist(data, bins);
axs[1].set_ylabel("observations", fontsize=14);
axs[2].hist(data, bins, density=False, weights=np.ones(len(data)) / len(data));
axs[2].set_ylabel("proportion", fontsize=14);
```
